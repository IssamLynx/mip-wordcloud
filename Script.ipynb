```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
import csv
balde =""
with open('MetaData_hybride,_assiduité,_étudiant.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
         balde += ', '.join(row)
```

```python
exclure_mots = ['d', 'du', 'de', 'la', 'des', 'le', 'et', 'est', 'elle', 'une', 'en', 'que', 'aux', 'qui', 'ces', 'les', 'dans', 'sur', 'l', 'un', 'pour', 'par', 'il', 'ou', 'à', 'ce', 'a', 'sont', 'cas', 'plus', 'leur', 'se', 's', 'vous', 'au', 'c', 'aussi', 'toutes', 'autre', 'comme']
wordcloud = WordCloud(background_color = 'white', stopwords = exclure_mots, max_words = 200).generate(balde)
print(wordcloud)
wordcloud.to_file('wordcloud.png')
plt.figure( figsize=(10,5) )
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
import yake
kw_extractor = yake.KeywordExtractor()
language = "fr"
max_ngram_size = 3
deduplication_threshold = 0.9
numOfKeywords = 200
custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)
keywords = custom_kw_extractor.extract_keywords(balde)
for kw,sc in keywords:
    if(sc>0.0006):
        print(kw, sc)
```
